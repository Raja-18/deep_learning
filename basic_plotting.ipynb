# Basic plotting

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```
x=np.random.random((10,1))
print(x)
```

## normal plot

```
plt.plot(x)                     # values of x are taken as x-axis and the indeces are taken as y-axis 
plt.show()
```

```
plt.plot(x,'*')             #plot only points excludes lines
plt.show()
```

```
plt.plot(x,'*-')              #plot both lines and points
plt.show()
```

```
x=np.linspace(0,10,100)            #it creates array of 100 elements starting from 0 and ends at 10
y=np.power(x,0.5)                  #it powers all elements of x with 0.5
plt.plot(x,y,'*')
plt.show()
```

## using seaborn

```
sns.set()
```

```
sns.lineplot(x,y)
plt.show()
```

## reading datas with pandas

```
datas=pd.read_csv('Toyota.csv')
datas.head()               # gives first 5 rows of the data
```

## filling nan values

```
datas['MetColor'].fillna(0,inplace=True)               # nan values is filled by 0 here but practically it should be filled using stats
```

```
datas.head()
```

## scatter plots

```
ax=sns.scatterplot(x='Price',y='Age',data=datas)
```

```
ax=sns.scatterplot(x='Price',y='Age',hue='Automatic',data=datas)   # with hue 
```

## histogram 

```
ax=sns.distplot(datas['Price'])
```

```
ax=sns.distplot(datas['Price'],kde=False,rug=True,bins=20)                # kde=False remove kernel density estimation
```

## box and whiskers plot

```
sns.boxplot(y='Price',data=datas)          #can find 1st quartile(25th percentile), median, 3rd quartile(75th percentile) and the whiskers q1-1.5*(IQR), q3+1.5*(IQR) and the outliers
```

```
sns.boxplot(x='MetColor',y='Price',data=datas)  # can compare catagorical datas (y-axis) with continuous data (x-axis)
```

```
sns.boxplot(x='CC',y='Price',data=datas)       # contineous data with contineous data
```

```
uniform_data=np.random.rand(10,12)              # array with 10 row and 12 columns
print(uniform_data)
```

## heatplot

```
ax=sns.heatmap(uniform_data)            # con find which number is intence in each cell of the array (10 rows - y axis) (12 columns - x axis)
```

```
ax=sns.heatmap(uniform_data,cmap="YlGnBu")                # can also change the colour
```
